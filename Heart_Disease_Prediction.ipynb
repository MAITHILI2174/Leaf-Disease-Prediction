Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Dataset



```python
dataset = pd.read_csv('/content/heart.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

Data Preprocessing

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
```

```python
print(x)
```

```python
print(y)
```

Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_test)
```

```python
print(Y_test)
```

Training Dataset

Training with Random Forest Classifier

```python
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train,Y_train)
```

```python
#SVM
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(X_train,Y_train)
```

```python
#kneighbors
from sklearn.neighbors import KNeighborsClassifier
model_kneighbors = KNeighborsClassifier()
model_kneighbors.fit(X_train,Y_train)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train,Y_train)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
model_random = RandomForestClassifier(n_estimators=200)
model_random.fit(X_train,Y_train)
```

```python
y_pred_logistic = model_logistic.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_kneighbors = model_kneighbors.predict(X_test)
y_pred_tree = model_tree.predict(X_test)
y_pred_random = model_random.predict(X_test)
```

Testing accuracy on different ML algorithm

```python
from sklearn.metrics import accuracy_score
logistic_Acc = accuracy_score(Y_test,y_pred_logistic)
svm_Acc = accuracy_score(Y_test,y_pred_svm)
kneignbors_Acc = accuracy_score(Y_test,y_pred_kneighbors)
Dectree_Acc = accuracy_score(Y_test,y_pred_tree)
random_Acc = accuracy_score(Y_test,y_pred_random)
```

```python
print(logistic_Acc)
print(svm_Acc)
print(kneignbors_Acc)
print(Dectree_Acc)
print(random_Acc)
```

Plotting Accuracy on different ML algorithms

```python
plt.figure(figsize=(12, 6))
plt.bar("Logistic Regression",logistic_Acc,width=0.5)
plt.bar("KNeighbors",kneignbors_Acc,width=0.5)
plt.bar("Support Vector Machine",svm_Acc,width=0.5)
plt.bar("Decision Tree",Dectree_Acc,width=0.5)
plt.bar("Random Forest",random_Acc,width=0.5)
plt.xlabel("Machine Learning Algorithm")
plt.ylabel("Accuracy")
```
